```python
def restar(x, y):
  return x - y
print(restar(4, 6))
```

```python
def saludar(nombre):
  print("Hola", nombre)
saludar("Galvis")
```

```python
def promedio(a, b, c):
  return (a+b+c)/3
print(promedio(1, 1, 1))
```

```python
notas = [2.5, 1, 3.8, 1.5]
promedio=sum(notas)/len(notas)
print("Su promedio de calificaciones es:", promedio)
print("Su resultado es: Aprobado" if promedio>=3 else "Su resultado es: Reprobado")
```

```python
ventas=[120000, 85000, 43000]
print("El promedio de ventas totales fue de:", sum(ventas)/len(ventas))
print("Cantidad de ventas", len(ventas))
for v in ventas:
  if v > 50000:
    print(v)
```

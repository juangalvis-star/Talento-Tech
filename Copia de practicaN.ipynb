```python
import pandas as pd

basepath = "/content/IRIS.csv"
df = pd.read_csv(basepath)
df.head()
```

```python
df.info()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns

sns.pairplot(df, hue='species')


```

```python
df.describe()
```

```python
# Crear un gráfico de barra para mostrar el promedio de longitud del sépalo por especie
sns.barplot(data=df, x='species', y='sepal_length')
plt.title('Promedio de Longitud del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Promedio de Longitud del Sépalo')
plt.grid(True)
plt.show()
```

```python
# Crear un grafico de linea para mostrar la variación de longitud de pétalo
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=df, x=df.index, y='petal_length', hue='species')
plt.title('variación de la Longitud del petalo por Especie')
plt.xlabel('Indice')
plt.ylabel('Longitud del Pétalo')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Diagrama de Dispersión: Longitud del sepalo vs Ancho del sepalo')
plt.xlabel('Longitud del sepalo')
plt.ylabel('Ancho del sepalo')
plt.legend(title='especie', loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataset

value_y_list = [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4] # valores de Y
value_x1_list = [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10] # valores de X1
value_x2_list = [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68] # valores de X2
value_x3_list = [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8] # valores de X3
value_x4_list = [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12] # valores de X4

data_dict = {"Y":value_y_list,
             "X1":value_x1_list,
             "X2":value_x2_list,
             "X3":value_x3_list,
             "X4":value_x4_list}

print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

#print(df) Tabla básica
```

```python
import numpy as np

# Convert the DataFrame to a NumPy array (matrix)
matrix_df = df.values

# Create a column of ones with the same number of rows as the matrix
ones_column = np.ones((matrix_df.shape[0], 1))

# Replace the first column of the matrix with ones
# We take all rows, the first column of ones, and then all columns from the second column of the original matrix onwards.
matrix_with_ones = np.concatenate((ones_column, matrix_df[:, 1:]), axis=1)

print("Original DataFrame as matrix:")
print(matrix_df)
print("\nMatrix with first column replaced by ones:")
print(matrix_with_ones)
```

```python
# Calculate the transpose of the matrix_with_ones
transposed_matrix = matrix_with_ones.T

# Display the transposed matrix
print("Transposed matrix:")
print(transposed_matrix)
```

```python
# Extract the 'Y' column as a NumPy array
y_array = df['Y'].values

# Reshape it into a column vector
y_vector = y_array.reshape(-1, 1)

print("Column vector Y:")
print(y_vector)
```

```python
# Perform matrix multiplication: transposed_matrix * matrix_with_ones
product_matrix = np.dot(transposed_matrix, matrix_with_ones)

print("Product of transposed_matrix and matrix_with_ones:")
print(product_matrix)
```

```python
# Calculate the inverse of the product_matrix
inverse_product_matrix = np.linalg.inv(product_matrix)

print("Inverse of the product_matrix:")
print(inverse_product_matrix)
```

```python
import numpy as np

# Perform the multiplication: (X'X)^-1 * X'
intermediate_product = np.dot(inverse_product_matrix, transposed_matrix)

# Perform the final multiplication: ((X'X)^-1 * X') * Y
beta_coefficients = np.dot(intermediate_product, y_vector)

print("Regression Coefficients (Beta):")
print(beta_coefficients)
```

```python
B0=beta_coefficients[0]
B1=beta_coefficients[1]
B2=beta_coefficients[2]
B3=beta_coefficients[3]
B4=beta_coefficients[4]
```

```python
df['Ypred'] = B0 + B1*df['X1'] + B2*df['X2'] + B3*df['X3'] + B4*df['X4']
df['Y-Ypred'] = df['Y'] - df['Ypred']
df['error'] = df['Y-Ypred']**2
df
```
